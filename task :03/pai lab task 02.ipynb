```python
def water_jug_dfs(capacity1, capacity2, target):

    stack = []
    visited = set()
    stack.append((0, 0, []))

    while stack:
        jug1, jug2, path = stack.pop()

        if (jug1, jug2) in visited:
            continue

        visited.add((jug1, jug2))

        # Check if target reached
        if jug1 == target or jug2 == target:
            print("\nTarget Achieved!\n")
            for step in path:
                print(step)
            print(f"Final State: ({jug1}, {jug2})")
            return


        next_states = []

        # 1. Fill Jug1
        next_states.append((capacity1, jug2, 
                           f"Fill Jug1  ({capacity1}, {jug2})"))

        # 2. Fill Jug2
        next_states.append((jug1, capacity2, 
                           f"Fill Jug2  ({jug1}, {capacity2})"))

        # 3. Empty Jug1
        next_states.append((0, jug2, 
                           f"Empty Jug1  (0, {jug2})"))

        # 4. Empty Jug2
        next_states.append((jug1, 0, 
                           f"Empty Jug2  ({jug1}, 0)"))

        # 5. Pour Jug1 → Jug2
        transfer = min(jug1, capacity2 - jug2)
        next_states.append((jug1 - transfer, jug2 + transfer,
                           f"Pour Jug1  Jug2 ({jug1 - transfer}, {jug2 + transfer})"))

        # 6. Pour Jug2 → Jug1
        transfer = min(jug2, capacity1 - jug1)
        next_states.append((jug1 + transfer, jug2 - transfer,
                           f"Pour Jug2 Jug1 ({jug1 + transfer}, {jug2 - transfer})"))

        # Add new states to stack (DFS behavior)
        for state1, state2, rule in next_states:
            if (state1, state2) not in visited:
                stack.append((state1, state2, path + [rule]))

    print(" No Solution Found.")

print("Water Jug Problem using DFS\n")

c1 = int(input("Enter capacity of Jug 1: "))
c2 = int(input("Enter capacity of Jug 2: "))
target = int(input("Enter target amount: "))

water_jug_dfs(c1, c2, target)
```
